```python
import tensorflow as tf

from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
model = models.Sequential([
    
    layers.Conv2D(
        64,
        (3,3),
        activation="relu",
        input_shape=(100,200,1)
    ),

    layers.MaxPooling2D(
        (2,2)
    ),

    layers.Conv2D(
        64,
        (3,3),
        activation="relu"
    ),

    layers.MaxPooling2D(
        (2,2)
    ),

    layers.Flatten(),

    layers.Dense(
        128,
        activation="relu"
    ),

    layers.Dense(
        31,
        activation="softmax"
    )
])
```
