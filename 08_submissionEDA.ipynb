```python
import pandas as pd
```

```python
df=pd.read_csv("submission.csv")
```

```python
print(df)
```

```python
wrong_ans = df[df["text"].str.len() < 6]
print(wrong_ans)
```

```python
same_length=df[df["text"].str.len() == 6]
print(same_length)
```

```python
lengths = df["text"].str.len().unique()
print(lengths)
```

```python
greater_len = df[df["text"].str.len() > 6]
print(greater_len)
```

```python
all_text = "".join(df["text"])
unique_chars = sorted(list(set(all_text)))

print(unique_chars)
print(len(unique_chars))
```

```python
import matplotlib.pyplot as plt, seaborn as sns

counts = df["text"].apply(list).explode().value_counts()
sns.barplot(x=counts.index, y=counts.values).set(title="Character Frequency")
plt.show()
```

```python
ax = sns.countplot(x=df["text"].str.len())
ax.set(title="Predicted Lengths")

for number in ax.containers:
    ax.bar_label(number)

plt.show()
```
