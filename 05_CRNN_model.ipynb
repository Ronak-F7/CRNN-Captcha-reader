```python
import tensorflow as tf

from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
inputs = layers.Input(
    shape=(100, 200, 1),
    name="image"
)
```

```python
x = layers.Conv2D(
    32,
    (3,3),
    activation="relu",
    padding="same"
)(inputs)

x = layers.MaxPooling2D(
    (2,2)
)(x)
```

```python
x = layers.Conv2D(
    64,
    (3,3),
    activation="relu",
    padding="same"
)(x)

x = layers.MaxPooling2D(
    (2,2)
)(x)
```

```python
x = layers.Reshape(
    target_shape=(50, 25 * 64)
)(x)
```

```python
x = layers.Bidirectional(
    layers.LSTM(
        128,
        return_sequences=True
    )
)(x)
```

```python
x = layers.Bidirectional(
    layers.LSTM(
        128,
        return_sequences=True
    )
)(x)
```

```python
x = layers.Dense(
    32,
    activation="softmax"
)(x)
```

```python
model = models.Model(
    inputs=inputs,
    outputs=x
)
```
